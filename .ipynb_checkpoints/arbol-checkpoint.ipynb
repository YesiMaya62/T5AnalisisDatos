```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

titanic_df = pd.read_csv("Titanic-Dataset.csv")
```

```python
#  An√°lisis Exploratorio de Datos
print("\n-----2. An√°lisis Exploratorio -----")
print(titanic_df.head())  # Primeras filas del DataFrame
print(titanic_df.info())  # Informaci√≥n general (tipos de datos, valores no nulos)
print(titanic_df.describe())  # Estad√≠sticas descriptivas (media, desviaci√≥n est√°ndar, etc.)

# Visualizaci√≥n de la Correlaci√≥n entre variables num√©ricas
numeric_cols = titanic_df.select_dtypes(include=np.number).columns
correlation_matrix = titanic_df[numeric_cols].corr()
plt.figure(figsize=(10, 8)) #Aumentamos el tama√±o de la figura
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlaci√≥n')
plt.show()


# Diagrama de dispersi√≥n para pares de variables num√©ricas
sns.pairplot(titanic_df[numeric_cols])
plt.show()

# Histograma de la variable objetivo ('Survived')
sns.histplot(titanic_df['Survived'])
plt.title('Distribuci√≥n de la Variable Objetivo (Survived)')
plt.show()
```

```python
# 3. Preprocesamiento de Datos

print("\n----- 3. Preprocesamiento de Datos -----")

# Define caracter√≠sticas num√©ricas y categ√≥ricas
numerical_features = ['Age', 'Fare', 'SibSp', 'Parch']
categorical_features = ['Sex', 'Embarked', 'Pclass']

# Pipeline de preprocesamiento
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='median'), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='drop'
)

# Crea un pipeline que combina preprocesamiento y el modelo de √Årboles de Decisi√≥n
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(random_state=42))
])
```

```python

# 4. Divisi√≥n del Conjunto de Datos (Train/Test)
print("\n----- 4. Divisi√≥n del Conjunto de Datos -----")
X_train, X_test, y_train, y_test = train_test_split(titanic_df.drop('Survived', axis=1), titanic_df['Survived'], test_size=0.2, random_state=42)
```

```python

# 5. B√∫squeda de Hiperpar√°metros (usando GridSearchCV)
print("\n----- 5. B√∫squeda de Hiperpar√°metros -----")
param_grid = {
    'classifier__max_depth': [None, 5, 10, 15],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print("Mejores hiperpar√°metros:", grid_search.best_params_)
best_model = grid_search.best_estimator_
```

```python

# 6. Entrenamiento del Modelo con los mejores hiperpar√°metros
print("\n----- 6. Entrenamiento del Modelo -----")
best_model.fit(X_train, y_train)
```

```python

# 7. Evaluaci√≥n del Modelo
print("\n----- 7. Evaluaci√≥n del Modelo -----")

# Predicciones en los conjuntos de entrenamiento y prueba
y_pred_train = best_model.predict(X_train)
y_pred_test = best_model.predict(X_test)


# M√©tricas de evaluaci√≥n
print("Conjunto de Entrenamiento:")
print(classification_report(y_train, y_pred_train))
print(confusion_matrix(y_train, y_pred_train))

print("\nConjunto de Prueba:")
print(classification_report(y_test, y_pred_test))
print(confusion_matrix(y_test, y_pred_test))


# 7. Visualizaci√≥n de Resultados (Matriz de Confusi√≥n)
print("\n----- 7. Visualizaci√≥n de Resultados -----")
cm = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No sobrevivi√≥', 'Sobrevivi√≥'],
            yticklabels=['No sobrevivi√≥', 'Sobrevivi√≥'])
plt.xlabel('Predicciones')
plt.ylabel('Valores Reales')
plt.title('Matriz de Confusi√≥n')
plt.show()
```

```python
# 11. Gr√°fica del √°rbol de decisi√≥n (si es peque√±o)
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 15)) #Aumenta el tama√±o para mejor visibilidad
plot_tree(best_model['classifier'], filled=True, 
          feature_names=best_model['preprocessor'].get_feature_names_out(), 
          class_names=['No sobrevivi√≥', 'Sobrevivi√≥'], rounded=True)
plt.title('√Årbol de Decisi√≥n')
plt.show()
```

```python

# 8. Interpretaci√≥n de Resultados (Similar a antes, pero con m√©tricas de clasificaci√≥n)
print("\n----- 8. Interpretaci√≥n de Resultados -----")
print("El modelo de √°rbol de decisi√≥n, despu√©s de la b√∫squeda de hiperpar√°metros, muestra un rendimiento aceptable tanto en el conjunto de entrenamiento como en el de prueba.  Sin embargo, se observa un desequilibrio en el recall, indicando que el modelo es m√°s preciso prediciendo la clase mayoritaria (no sobrevivientes).  Esto se puede mejorar mediante t√©cnicas de manejo de desequilibrio de clases como sobremuestreo o submuestreo, o usando m√©tricas de evaluaci√≥n m√°s adecuadas para datos desbalanceados como el AUC-ROC.")
print("La visualizaci√≥n del √°rbol proporciona una visi√≥n clara de las reglas de decisi√≥n utilizadas por el modelo para predecir la supervivencia. Se pueden observar las variables m√°s importantes que influyen en la predicci√≥n y c√≥mo se combinan para generar las decisiones.")
```
